Exploratory Data Analysis

Below we visualize the distributions and relationships in the dataset to understand how features relate to student performance.

```python
df.groupby("race/ethnicity")[["math score", "reading score", "writing score"]].mean().round(2)

avg_scores = (
    df.groupby("race/ethnicity")[["math score", "reading score", "writing score"]]
      .mean()
      .round(2)
      .reset_index()
)

avg_scores
```
